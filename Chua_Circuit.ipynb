In this notebook, we will construct the chaotic Chua electric circuit system whose circuit equations are given as follows:
\begin{align}
\frac{dV_1}{dt}=&\frac{G(V_2-V_1)-f(V_1)}{C_1}\\
\frac{dV_2}{dt}=&\frac{G(V_1-V_2)+i}{C_2}\\
\frac{di}{dt}=&-\frac{V_2}{L}\\
G=&R^{-1},
\end{align}
where $V_1,V_2,I$ are the voltages across $C_1, C_2$ and the current passing by $L$. The non-linear resistor $N_R$ registers $f(V_1)=G_bV_1+0.5(G_a-G_b)(|V_1+B_p|-|V_1-B_p|)$.

Converting the Chua's circuit equatoin into following dimensionless form:
\begin{align}
\dot{x}=&a(y-x-cx-\frac{1}{2}(d-c)(|x-1|-|x+1|))\\
\dot{y}=&x-y+z\\
\dot{z}=&-by.
\end{align}



```python
from matplotlib import pyplot as plt
from matplotlib.animation import ArtistAnimation
import numpy as np
from scipy.integrate import odeint
```

```python
fig=plt.figure()
```

```python
mo=-1/np.pi
m1=2/np.pi
tmax=10;
```

```python
def chua(x,t):
  return [a*(x[1]-(m1*x[0]+(mo-m1)/2*(np.abs(x[0]+1)-\
                                      np.abs(x[0]-1)))), x[0]-x[1]+x[2],-15*x[1]]
```

```python
time=np.arange(0,tmax,.1)
x0=[1.96,-.05619*np.exp(-9),-3.077]
myimages=[]
for a in np.arange(8,11,0.1):
  xs=odeint(chua,x0,time)
  imgplot=plt.plot(xs[:,0],xs[:,1],"r-")
  myimages.append(imgplot)

my_anim=ArtistAnimation(fig,myimages,interval=50,\
                        blit=False,repeat_delay=500)
plt.show()
```
