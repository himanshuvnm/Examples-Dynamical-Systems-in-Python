

We will be displaying the phase potrait of a nonlinear system with emphasis on the vector plot. The nonlinear dynamical system taken into the present consideration is described as follows:
\begin{align}
\dot{x}=&y\\
\dot{y}=&x(1-x^2)+y.
\end{align}



```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
import pylab as pl
```

Now, we will provide the differential equaiton as follows.

```python
def dx_dt(x,t):
  return [x[1],x[0]*(1-x[0]**2)+x[1]]
```

Now, we will generate the trajectories in the forward time.

```python
ts=np.linspace(0,10,500)
ic=np.linspace(-3,3,6)
for r in ic:
  for s in ic:
    x0=[r,s]
    xs=odeint(dx_dt,x0,ts)
    plt.plot(xs[:,0],xs[:,1],"r-")
```

The following generates the trajectories in the backward time. Then also, we have the full phase potrait of the above dynamical system.

```python
ts=np.linspace(0,-10,500)
ic=np.linspace(-3,3,6)
for r in ic:
  for s in ic:
    x0=[r,s]
    xs=odeint(dx_dt,x0,ts)
    plt.plot(xs[:,0],xs[:,1],"r-")

plt.xlabel("x",fontsize=15)
plt.ylabel("y",fontsize=15)
plt.tick_params(labelsize=15)
plt.xlim(-3,3)
plt.ylim(-3,3);

X,Y=np.mgrid[-3:3:20j,-3:3:20j]
u=Y
v=X*(1-X**2)+Y
pl.quiver(X,Y,u,v,color='b')
plt.show()
```

In the present example, we will produce the phase potrait of Capacitor-Resistor Electrical Circiut system whose governing equations are given as:
\begin{align}
\dot{x}=&y\\
\dot{y}=&-x+x^3-(a_0+x)y
\end{align}
Let's start the analysis now.

```python
a0=0.01
```

```python
def dx_dt(x,t):
  return [x[1],-x[0]+x[0]**3-(a0+x[0])*x[1]]
```

```python
ts=np.linspace(0,3,50)
ic=np.linspace(-3,3,6)
for r in ic:
  for s in ic:
    x0=[r,s]
    xs=odeint(dx_dt,x0,ts)
    plt.plot(xs[:,0],xs[:,1],"r-")

plt.xlabel("x",fontsize=15)
plt.ylabel("y",fontsize=15)
plt.tick_params(labelsize=15)
plt.xlim(-3,3)
plt.ylim(-3,3);

X,Y=np.mgrid[-3:3:20j,-3:3:20j]
u=Y
v=-X+X**3-(a0+X)*Y
pl.quiver(X,Y,u,v,color='b')
plt.show()
```
