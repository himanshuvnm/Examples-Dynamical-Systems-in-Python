In this notebook, we will consider several examples of non-autonomoues dynamical system. We bign first by working on the phase portraits and the poincare maps of the Duffing oscillator system whose governing equation is given as follows:
\begin{align}
\ddot{x}+k\dot{x}+\beta x+\alpha x^3=Γ\cdot\cos\left(\omega t\right).
\end{align}
This can be augmented to following set of first order differential equations:
\begin{align}
\dot{x}=&y\\
\dot{y}=&-\beta x-ky-\alpha x^3+Γ\cdot\cos\left(\omega t\right).
\end{align}

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

```python
xmin=-2
xmax=2
ymin=xmin
ymax=xmax
```

```python
k=0.3
omega=.25
gamma=.5
```

```python
def dx_dt(x,t):
  return [x[1],x[0]-k*x[1]-x[0]**3+gamma*np.cos(omega*t)]
```

```python
# Phase portrait of Duffing system
t=np.linspace(0,500,10000)
xs=odeint(dx_dt,[1,0],t)
plt.plot(xs[:,0],xs[:,1],"r-",lw=.5)
plt.xlabel('x',fontsize=18)
plt.ylabel('y',fontsize=18)
plt.tick_params(labelsize=18)
plt.grid()
plt.xlim(xmin,xmax)
plt.ylim(ymin,ymax);
plt.title('Phase Portrait')
```

```python
# The Poincare maps of the Duffing system
x=[]
y=[]
fig, ax=plt.subplots(figsize=(6,6))
t=np.linspace(0,4000*(2*np.pi)/omega,16000000)
xs=odeint(dx_dt,[1,0],t)
x=[xs[4000*i,0] for i in range(4000)]
y=[xs[4000*i,1] for i in range(4000)]

ax.scatter(x,y,color='blue',s=.1)
plt.xlabel('x',fontsize=18)
plt.ylabel('y',fontsize=18)
plt.tick_params(labelsize=16)
plt.title('Poincaré section')
plt.grid()
plt.show()
```

```python
def dx_dt(x,t):
  return [-x[1]+k_const*x[0]*x[1]*np.sin(omega*t)+t,-x[0]+k_const*x[1]*np.cos(omega*t)]
```

```python
t=np.linspace(0,500,1000000)
xs=odeint(dx_dt,[1,0],t)
plt.plot(xs[:,0],xs[:,1],"r-",lw=.5)
plt.xlabel('x',fontsize=18)
plt.ylabel('y',fontsize=18)
plt.tick_params(labelsize=18)
plt.grid()
plt.xlim(xmin,xmax)
plt.ylim(ymin,ymax);
plt.title('Phase Portrait')
```

```python
# The Poincare maps of the Duffing system
x=[]
y=[]
fig, ax=plt.subplots(figsize=(6,6))
t=np.linspace(0,4000*(2*np.pi)/omega,160000)
xs=odeint(dx_dt,[1,0],t)
x=[xs[400*i,0] for i in range(400)]
y=[xs[400*i,1] for i in range(400)]

ax.scatter(x,y,color='blue',s=.1)
plt.xlabel('x',fontsize=18)
plt.ylabel('y',fontsize=18)
plt.tick_params(labelsize=16)
plt.title('Poincaré section')
plt.grid()
plt.show()
```
